### Naive Bayes Classifier:

Bayes’ Theorem can calculate the probability of a piece of data belonging to a given class, given our prior knowledge. Bayes’ Theorem is stated as:

P(class|data) = (P(data|class) * P(class)) / P(data)

Where P(class|data) is the probability of class given the provided data.

```python
import pandas as pd
import numpy as np
import math

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_csv('iris.csv',header=None)
print('Rows',df.shape[0],'Columns',df.shape[1])
# print('\nTypes:\n\n',df.dtypes)
df.head(3)
```

```python
sns.pairplot(df,hue=4);
```

### Split the dataset into 75% training set and 25% test set:

```python
df, test = np.split(df.sample(frac=1, random_state=42), [int(.75*len(df))])
```

### Change target string column to int:

```python
dicco = {}
for i,j in enumerate(df[4].unique()):
    dicco[j] = i
df[4] = df[4].apply(lambda x: dicco[x]) 
dicco
```

### Separate classes:

```python
diccy = {}
for index,row in df.iterrows():
    ## create a vector for each row (minus the target info)
    vector = list(row)[:-1]
    class_value = list(row)[-1]
    if class_value not in diccy:
        diccy[class_value] = []
    diccy[class_value].append(vector)
# diccy
```

### Create summary of stats by class for each columns:

```python
def summarize_dataset(data):
    summaries = [(np.mean(col), np.std(col), len(col)) for col in zip(*data)]
    return summaries

summary = {}
for class_value, rows in diccy.items():
    summary[class_value] = summarize_dataset(rows)
# summary
```

### Gaussian Probability Density Function (Gaussian PDF)

Calculating the probability or likelihood of observing a given real-value is to assume that these real values are drawn from a distribution, such as a bell curve or Gaussian distribution.

A Gaussian distribution can be summarized using only two numbers: the mean and the standard deviation. Therefore, we can estimate the probability of a given value. This is called a Gaussian Probability Distribution Function and can be calculated as:

f(x) = (1 / sqrt(2 * PI) * sigma) * exp(-((x-mean)^2 / (2 * sigma^2)))

Where sigma is the standard deviation for x, mean is the mean for x and PI is the value of pi.

```python
def plot_probabilities(probabilities):
    sns.barplot(x=list(dicco.keys()), y=list(probabilities.values()))
    plt.show()
    
def predict(summaries, row, plot=True):
    probabilities = {}
    number_of_rows = sum([summaries[label][0][-1] for label in summaries])
    for class_value, class_summary in summary.items():
        ## number of elements in each class / total number of elements
        probabilities[class_value] = summary[class_value][0][2]/float(number_of_rows)
        ## for each column within each class:
        for i in range(len(class_summary)):
            mean, std, _ = class_summary[i]
            ## Calculate the Gaussian probability distribution function for x
            probabilities[class_value] *= (1 / (np.sqrt(2 * np.pi) * std)) * np.exp(-((row[i]-mean)**2 / (2 * std**2 )))

    if plot==True:
        plot_probabilities(probabilities)
    
    best_label, best_prob = None, -1
    for class_value, probability in probabilities.items():
        if best_label is None or probability > best_prob:
            best_prob = probability
            best_label = [[key  for (key, value) in dicco.items() if value == class_value],class_value]
    return best_label
```

### Predict one record:

```python
## New flower:
row = [6.6,3,5.5,2]

flower = predict(summary, row, plot=True)
print('New record={}, Prediction: {}'.format(row, flower[0][0]))
```

### Predict on the test set:

```python
sum_true = 0
for index,row in test.iterrows():
    vector = list(row)[:-1]
    pred = predict(summary, vector, plot=True)
    print('Actual label:',list(row)[-1],'prediction:',pred[0][0])
    if list(row)[-1] == pred[0][0]:
        sum_true += 1
print('Accuracy:',np.round(sum_true/len(test)*100,3))
```

```python
df_pred = pd.DataFrame()
actual_labels=[]
pred_labels=[]

for index,row in test.iterrows():
    vector = list(row)[:-1]
    
    actual_labels.append(list(row)[-1])
    pred_labels.append(predict(summary, vector, plot=False)[0][0])

df_pred['actual_labels'] = actual_labels
df_pred['pred_labels'] = pred_labels


def highlight_diff(df_pred):
    r = 'red'

    diff = df_pred['actual_labels'] != df_pred['pred_labels']

    df1 = pd.DataFrame('background-color: ', index=df_pred.index, columns=df_pred.columns)
    df1['pred_labels'] = np.where(diff,'background-color: {}'.format(r), df1['pred_labels'])
    print('Accuracy:',np.round(sum(~diff)/len(diff)*100,3))
    return df1


df_pred.style.apply(highlight_diff, axis=None)
```
